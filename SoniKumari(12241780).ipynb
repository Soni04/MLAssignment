```python
import numpy as np
import matplotlib.pyplot as plt

# Defined the data
x1 = np.array([1, 1, 0, 5, 6, 4])
x2 = np.array([4, 3, 4, 1, 2, 0])
x = np.column_stack((x1, x2))

# Plot the data
plt.scatter(x[:, 0], x[:, 1], c='blue', marker='o')

# Set labels and other plot settings
plt.xlabel(r'$X_1$')
plt.ylabel(r'$X_2$')
plt.title('Scatter Plot of Observations')

# Annotate the points with their indices
for i, (x_val, y_val) in enumerate(x, start=1):
    plt.text(x_val + 0.15, y_val, str(i), fontsize=9)

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Final cluster 
clusters = np.array([1, 1, 1, 2, 2, 2])  

# Defined colors for clusters
colors = {1: 'red', 2: 'blue'}

plt.figure(figsize=(8, 6))
for cluster in np.unique(clusters):
    cluster_points = x[clusters == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=colors[cluster], label=f'Cluster {cluster}', edgecolor='k', s=100)

# Add labels and title
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Observations Colored with respect to Cluster')
plt.legend()
plt.grid(True)
plt.show()
```
